```python
import joblib
import pandas as pd
```

```python
model = joblib.load("../model.pkl")
encoder = joblib.load("../encoder.pkl")

print("Model & Encoder Loaded!")
```

```python
print(encoder.classes_)
```

```python
def predict_traffic(hour, day, month, temp, weather):
    weather_encoded = encoder.transform([weather])[0]

    input_data = pd.DataFrame([{
        'hour': hour,
        'day': day,
        'month': month,
        'temp': temp,
        'weather_main': weather_encoded
    }])

    # Prediction
    prediction = model.predict(input_data)[0]

    return prediction
```

```python
pred = predict_traffic(
    hour=8,
    day=2,
    month=5,
    temp=25,
    weather='Clear'
)

print("Predicted Traffic Volume:", pred)
```
